```python
#I put the files I found into a dict that I could use for a for loop to createa data frame using pandas. This step made the rest of the
#analytics much more efficient
d1 = {'url': r'C:\Users\pad39\Downloads\all-ages (6).csv',
     'name': 'college majors',
     'file_name': 'all-ages (6).csv',  
     'load_func': pd.read_csv}

d2 = {'url': r'C:\Users\pad39\Downloads\games\metacritic_games.json',
     'name': 'meta critic games',
     'file_name': 'metacritic_games.json',
     'load_func': pd.read_json}
     
d3 = {'url': r'C:\Users\pad39\Downloads\forestfires.xlsx',
     'name': 'forest fires',
     'file_name': 'forestfires.xlsx',       
     'load_func':pd.read_excel}

df_dict = [d1, d2, d3]
```

```python
#takes the elements of he dicts listed above and appends them to the lists below using a for loop that iterates through the dicts
name = []
source = []
num_rows = []
num_columns = []
file_name = []
for info in df_dict:
    info['data'] = info['load_func'](info['url'])
    rows, columns = info['data'].shape
    num_rows.append(rows)
    num_columns.append(columns)
    source.append(info['url'])
    name.append(info['name'])
    file_name.append(info['file_name']) 
df = pd.DataFrame({'names': name, 'file_name': file_name, 'source': source, 'num_rows': num_rows, 'num_columns': num_columns})
df.to_csv

```
