```python
!pip install sklearn
import pandas as pd
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import GaussianNB
```

# Dataset 1

```python
df1 = pd.read_csv('data/dataset1.csv')
sns.pairplot(data =df1, hue='char')
```

```python
# I do not think the Gaussian Naive Bayes to work well for this data set because the features overlap a lot which means they may be 
#not be independent of each other
```

```python
gnb=GaussianNB()
```

```python
df1.values
```

```python
df1.values[:,1:5]
```

```python
x_train, x_test, y_train, y_test = train_test_split(df1.values[:,1:5],
                                                   df1.values[:,-1],
                                                   test_size=.5,
                                                   random_state=0)
```

```python
gnb.fit(x_train, y_train)
```

```python
y_pred=gnb.predict(x_test)
y_pred
```

```python
sum(y_pred == y_test)
```

```python
len(y_pred)
```

```python
#The Guassian Naive Bayes predicted 91 of the 100 classifications 
#correctly. 91% is good but not great for a machine learning
#model as I said the features of the data set seemd clustered
#together which leads me to believe that They are not Indipendent
#If this is true than maybe difficult to identify a good classifier 
#because if they all go up at the same time it gets tricky 
#to guess which letter it would be
```
