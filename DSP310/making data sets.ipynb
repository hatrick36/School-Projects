```python

spotify17=pd.read_sql_query('SELECT * FROM featuresdf',con)
```

```python
spotify17.head()
```

```python
#I used the iloc function to shift the first row of the df up and then the drop function to get rid of the orgional header
spotify17=spotify17.rename(columns=spotify17.iloc[0]).drop(spotify17.index[0])
```

```python
spotify17.shape
```

```python
#I concatted the two dataframes to one large one so as you can 
#see from the shape it has double the rows but the same amount of 
#columns because we simply put one df under the other
spotify_df=pd.concat([spotify17,spotify18])
spotify_df.shape
```

```python
spotify_df
```

```python
spotify_df.columns
```

```python
spotify_merge=pd.merge(spotify17,spotify18,left_on='artists',right_on='artists')
spotify_merge.head(2)
```

```python
#unlike the concatted df this dataframe was merged there fore 
#all columns for the two are still seperate
spotify_merge.shape
```

```python
#when merging the data frames entries with artists that are in both data frames
#we find 95 rows in this merge as opposed to 193 rows with the "outer" comand.
pd.merge(spotify17,spotify18,on='artists',how='inner')
```

```python
#I added a column to the data set that counts the number of artists
spotify_df['artist_count']=spotify_df.artists.count()
spotify_df
```

```python
# for some reason the columns in the merge data set were strings so i had to set them a numeric to calculate their averages
energy_x=pd.to_numeric(spotify_merge['energy_x'])
energy_y=pd.to_numeric(spotify_merge['energy_y'])
```
